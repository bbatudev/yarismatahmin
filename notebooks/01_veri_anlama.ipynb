```
import pandas as pd
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Dosya yollar?n? tan?mlama
data_dir = '../march-machine-leraning-mania-2026'
csv_files = glob.glob(os.path.join(data_dir, '*.csv'))
print(f"Toplam {len(csv_files)} adet CSV dosyas? bulundu.")
```

```
# T?m dosyalar?n temel sat?r-s?tun bilgileri ve eksik veri durumlar?n? ?zetleme
summary = []
for f in csv_files:
    try:
        df = pd.read_csv(f)
        missing = df.isnull().sum().sum()
        summary.append({
            'Dosya Ad?': os.path.basename(f),
            'Sat?r Say?s?': df.shape[0],
            'S?tun Say?s?': df.shape[1],
            'Eksik De?er Say?s?': missing
        })
    except Exception as e:
        print(f"Hata: {os.path.basename(f)} - {e}")

summary_df = pd.DataFrame(summary).sort_values(by='Sat?r Say?s?', ascending=False).reset_index(drop=True)
display(summary_df.head(10))
```
